```python
import pandas as pd
df_r = pd.read_csv('C:/F/我的坚果云/Alien Jaeger 2019/02 专栏相关/ltv/retention_data.csv',engine='python')
df_r
```

```python
df_r2 = pd.melt(df_r,id_vars=['Date'])
#df_r2
df_r2['x'] = df_r2['variable'].str.split(' ',1,True)[1].astype(int)
df_r2['y'] = df_r2['value'].str.split('%',1,True)[0].astype(float)/100
df_r2[0:10]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(x=df_r2['x'],y=df_r2['y'])
```

```python
df_a = pd.read_csv('C:/F/我的坚果云/Alien Jaeger 2019/02 专栏相关/ltv/arpu_data.csv',engine='python')
df_a
```

```python
df_a2 = pd.melt(df_a,id_vars=['Date'])
df_a2['x'] = df_a2['variable'].str.split(' ',1,True)[1].astype(int)
df_a2[0:10]
```

```python
plt.scatter(x=df_a2['x'],y=df_a2['value'])
```

```python
from scipy.optimize import curve_fit
import numpy as np
```

```python
def fun_r(x, a, b):
    return a**x + b
popt1, pcov1 = curve_fit(fun_r, df_r2['x'], df_r2['y'])
y1 = [fun_r(i,popt1[0],popt1[1]) for i in df_r2['x']]
plt.plot(df_r2['x'],y1,'r--')
```

```python
def fun_a(x, a, b, c):
    return a*x*x + b*x + c
popt2, pcov2 = curve_fit(fun_a, df_a2['x'], df_a2['value'])
y2 = [fun_a(i,popt2[0],popt2[1],popt2[2]) for i in df_a2['x']]
plt.plot(df_a2['x'],y2,'r--')
```

```python
s = 0
for i in range(1,18):
    s = s + fun_r(i,popt1[0],popt1[1])*fun_a(i,popt2[0],popt2[1],popt2[2])
print(s)
```
